```python

def draw_correlation_heatmap(dataframe):
    # Calculate the correlation matrix
    corr_matrix = dataframe.corr()

    # Set up the matplotlib figure
    plt.figure(figsize=(10, 8))

    # Create the heatmap using Seaborn
    sns.heatmap(corr_matrix,  cmap='coolwarm', fmt='.2f')

    # Add title
    plt.title('Correlation Heatmap')

    # Display the plot
    plt.show()
```

```python
# Call the function to draw the correlation heatmap
draw_correlation_heatmap(df1)
```

```python
# Function to filterout the correlated features
def correlation(dataset, threshold):
    col_corr=set()
    corr_matrix=dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            
            if abs(corr_matrix.iloc[i,j])>threshold:
                colname=corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
    
```

```python
corr_features=correlation(df1, 0.85)
len(corr_features)
```

```python
corr_features
```

# Conclusion-Above are the top 10 features which are highly correlated

```python
df_corr=df1[corr_features]
```

```python
df_corr
```

```python
draw_correlation_heatmap(df_corr)
```
